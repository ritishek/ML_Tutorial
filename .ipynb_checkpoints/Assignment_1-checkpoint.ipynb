```python
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm,stats
```

```python
import numpy as np
```

```python
x1=np.random.normal(5,1,10)
plt.hist(x1,10,density=True)
plt.plot(x1,norm.pdf(x1,5,1))
plt.show()
```

```python
sigma,mu=5,1
s = np.random.normal(mu, sigma, 10)
count, bins, ignored = plt.hist(s, 10, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```

```python
import math
mu = 5
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 1.60*sigma, mu + 1.60*sigma,10)
y=norm.pdf(x,mu,sigma)
plt.plot(x,y,'-o',color='b',linestyle=' ')
plt.title('Gaussian Distribution of 10 samples with μ = 5 and σ = 1')
#plt.show()
plt.savefig('gaussian.png')
```

```python
y1=norm.pdf(x,5,1)
y1
```

```python
fig,ax=plt.subplots(2,5,figsize=(20,6))
```

```python
fig,ax=plt.subplots(3,4,figsize=(25,10))
for mu in range(12):
    y=norm.pdf(x,mu,1)
    plt.subplot(3,4,mu+1)
    plt.plot(x,y,color='r')
    plt.plot(x,y1,'-o',color='b',linestyle=' ')
    plt.xlim(3,7)
    plt.ylim(0.00,0.50)
    plt.text(6.1,0.45,'μ = '+str(mu)+', σ = 1',color='g')
#plt.show()
plt.savefig('mle.png')
```
